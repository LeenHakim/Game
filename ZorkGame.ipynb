```python
#define a class for player which has health and name of the player/character
class Character:
    #5 params
    #3 as integer name, health, time
    #2 boolean rope, water
    def __init__ (self,name,hp,time,rope, water):
        self.health=hp
        self.name= name
        self.time=time
        self.rope = rope
        self.water = water
    
    #getters and setters
    def get_health(self):
        return self.health
    
    def get_name(self):
        return self.name
    
    def get_time(self):
        return self.time
    
    def set_health(self,new_hp):
        #to set new health value we minus the new hp value from the original hp
        self.health -=new_hp
        #if hp is less than or equal 0 -> game over
        if (self.health <= 0):
            print('%s has died!' %self.name)
        #if new hp value is negative then it will be added to the original hp
        elif(new_hp<0):
            print('%s has gained %i HP!' % (self.name, -new_hp))
            print("\n")
        #if new hp is positive it will be reduced from the original hp
        elif(new_hp>=0):
            print('%s has lost %i HP!' % (self.name, new_hp))  
            print("\n")

    def set_name(self,new_name):
        self.name=new_name
  
    def set_time(self,new_time):
        #to set new time we reduce the new time value from the original time 
        self.time -=new_time
        #if original time is less than or equal 0 -> game over
        if(self.time<=0):
            print('Your time is over!')
        else:
            print('%s has spent %.2f hour of their time' %(self.name,new_time))
            print("\n")
    
    def get_rope(self):
        return self.rope

    def get_water(self):
        return self.water

    def set_rope(self, new_rope):
        self.rope = new_rope

    def set_water(self, new_water):
        self.water = new_water
  
  #function to print a summary of the player's current status
    def character_summary(self):
        print ( 
           f"Your current status is: \n Remaining HP: {self.health} \n Remaining Time: {self.time} \n"
        )
     
 #function that will be called when the game starts to introduce the game and let player insert the name
def character_select():
    print("This is a survival zork game! your mission is to rescue a girl who has been missing!\n She is a pharmacist and she is from your village 'Easthaven'!\n She got lost in the forest while she was on her way to deliver medicine to the nearby village 'Northon' \n ")
    character_name = input('What is your name? \n')
    print('WELCOME',character_name, '! \n')
    print("You will start with 100 HP and 12 hours to rescue the girl, good luck! \n")  
    return (character_name)
```

```python
#define class weapon which has a name and id
class Weapon:
    def __init__ (self,name,id):
        self.name=name
        self.id=id
    #getters and setters
    def get_name(self):
        return self.name
    
    def set_name(self,new_name):
        self.name=new_name
    
    def get_id(self):
        return self.id
    
    def set_id(self,new_id):
        self.id=new_id

#Object of the Weapon class 
player_weapon = Weapon(name=None, id=None)

#selecting weapon function
def weapon_select():
    #player inserts their weapon choice 1,2 or 3
    weapon_selection= input('Choose your weapon: Machete (1) \n or Gun (2) \n or Bow and arrow (3) ? \n')
    #if the selection input wasn't 1,2 or 3 it will ask to input a correct value
    while weapon_selection!='1' and weapon_selection!='2' and weapon_selection!='3':
        print('Invalid! Choose either 1 or 2 or 3!')
        weapon_selection= input('Choose your weapon: Machete (1) \n or Gun (2) \n or Bow and arrows (3) ? \n')
        
    #setting weapon name and id based on the input
    if weapon_selection=='1':
        player_weapon.set_name('Machete')
        player_weapon.set_id(1)
        print('Now you have a machete in your inventory \n')
        return(player_weapon)
    elif weapon_selection =='2':
        player_weapon.set_name('Gun')
        player_weapon.set_id(2)
        print('Now you have a gun in your inventory \n')
        return(player_weapon)
    elif weapon_selection=='3':
        player_weapon.set_name('Bow and arrows')
        player_weapon.set_id(3)
        print('Now you have a bow and arrows in your inventory \n')
        return(player_weapon)
```

```python
#class path with direction param
class Path:
    def __init__ (self,direction):
        self.direction=direction
    
    #getters and setters 
    def get_direction(self):
        return self.direction
    
    def set_direction(self,new_direction):
        self.direction=new_direction

#function to let player select path
def path_select():
    #inserts path 1 or 2
    path_selection= input('Choose your direction: East (1) \n or West (2)? \n')
    #if not 1 or 2 inserted show error and ask again to insert
    while path_selection!='1' and path_selection!='2':
        print('Invalid! Choose either 1 or 2!')
        path_selection= input('Choose your direction: East (1)  \n or West (2)? \n')
    #if 1 -> set the direction of the object eat_path as east and return it
    if path_selection=='1':
        print('You have chosen East! it is the path that is used by all the people crossing the forest between the two villages. \n')
        east_path = Path('east')
        return east_path.direction
    #if 2 -> set the direction of the object west_path as west and return it
    elif path_selection=='2':
        print('You have chosen West! This is a mesterious and dangerous path which nobody has crossed for ages! \n')
        west_path= Path('west')
        return west_path.direction
```

```python
#fifth obstacle
def encounter_snake():
    #let player insert their choice
    print("on your journey, you realize that there is something moving, you look closely and you see a snake! Now you will need to use your weapon to kill the snake")
    choice= input("Press (1) in order to kill the snake")
    #if choice is not ask player to insert again until it's 1
    while choice!='1':
        print("Invalid! Sorry you don't have another option but to kill the snake please press 1!")
        choice= input("Press (1) to kill the snake!")
    #if choice is 1 check weapon and if girl has water to set new time and health
    if(choice=='1'):
        if player_weapon.get_id()==1:
            if(girl.get_water()==False):
                character.set_time(1)
                character.set_health(30)
            else:
                character.set_health(30)

        elif player_weapon.get_id()==2:
            if(girl.get_water()==False):
                character.set_time(3)
                character.set_health(50)
            else:
                character.set_health(50)
    
        elif player_weapon.get_id()==3:
            if(girl.get_water()==False):
                character.set_time(4)
                character.set_health(70)
            else:
                character.set_health(70)
```

```python
#in case the player ignored the water or didn't go through that path to find it in the first place
def find_water():
    #checks that player doesn't have water
    if (character.get_water()==False):
        #let player input ok to continue
        choice= input("You have to look for water to bring the girl! Let's go! (ok)")
        #if he doesn't keep asking
        while choice!='ok':
            print("Invalid! Sorry you don't have another option but finding the water for now so please press (ok)")
            choice = input("You have to look for water to bring the girl! Let's go! (ok)")
        #once choice is ok set new time,water for both character and girl to true
        if(choice=='ok'):
            character.set_health(0)
            character.set_time(1)
            character.set_water(True)
            girl.set_water(True)
            print("You found the water and luckily nothing stepped in your way, now you're on your way back to give the water to the girl")
  
```

```python
#function to allow player to gain health
def rest():
    #take player choice 1 or 2
    choice=input("Would you like to rest? you will spend some time resting but it will be worth it! yes (1) \n or no (2)")
    #keep asking if choice not 1 or 2
    while choice!='1' and choice !='2':
        print('Invalid! Choose either 1 or 2!')
        choice= input("Do you want to rest? yes (1) \n or no? (2)")
    #if choice is 1 check if girl has water and set new health and time based on that
    if(choice=='1'):
        if(girl.get_water()==False):
            character.set_health(-30)
            character.set_time(3)
        #if girl has water there's no need to keep counting time so we don't set it
        else:
            character.set_health(-30)
    elif(choice=='2'):
        print("Ok then good luck with the rest of your journey!")
   
```

```python
#after tiger obstacle the player finds water on their way 
def take_water():
    #let player choose whether to take water or not 1 or 2
    choice= input("You find water on your left side but you don't need it! Do you want to take the water with you anyways? (1) \n or ignore it? (2)")
    #if not 1 or 2 entered keep asking
    while choice!='1' and choice !='2':
        print('Invalid! Choose either 1 or 2!')
        choice= input("Do you want to take the water with you anyways? (1) \n or ignore it? (2)")
    #if 1 set water to true
    if(choice=='1'):
        character.set_water(True)
        print("Fair enough, now you have water with you!")
    #if 2 continue journey
    else:
        print("Let's keep going then!")
```

```python
#last function to be called when the player has finished his journey
def arrive_at_village():
    print("After a long journey, you have finally arrived at the village with the girl!")
    print("\n Congratulations on finishing the game, you've done a great job %s!" %character.get_name())
    print("\n See you soon!")
```

```python
#function to restart the game
def restart():
    #ask if player wants to restart
    choice= input("want to restart the game? (yes) or (no)")
    while choice!='yes' and choice !='no':
        print('Invalid! Choose either yes or no!')
        choice= input("want to restart the game? (yes) or (no)")
    if choice=='yes':
        start_game()
    else:
        print("Ok see you soon :)")
    
```
